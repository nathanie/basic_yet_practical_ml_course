```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('../input/KaggleV2-May-2016.csv')
```

first lets look into our data, we'll check its size and structure and come up with some questions about it

lets look at the features we have got:

```python
pd.DataFrame({'feature':data.columns,'unique_values':[len(data[f].unique()) for f in data.columns]})
```

```python
data = pd.concat([data,pd.get_dummies(data.Handcap,prefix='Handcap',prefix_sep='_')],axis=1)
data = pd.concat([data,pd.get_dummies(data.Neighbourhood,prefix='Neighbourhood',prefix_sep='_')],axis=1)

data.drop(['Handcap','Neighbourhood'],axis=1,inplace=True)
data.Gender = data.Gender.replace(to_replace=['F','M'],value=[1,0])
```

```python
temp = pd.to_datetime(data.ScheduledDay)
time_features_schedule = pd.DataFrame({'dow_sch':temp.dt.dayofweek,'hour_sch':temp.dt.hour,'day_sch':temp.dt.day})
temp = pd.to_datetime(data.AppointmentDay)
time_features_appointment = pd.DataFrame({'dow_app':temp.dt.dayofweek,'hour_app':temp.dt.hour,'day_app':temp.dt.day})
data = pd.concat([data,time_features_schedule,time_features_appointment-time_features_schedule.values],axis=1)
del temp
```

```python
data.tail()
```

```python
omit_columns = ['AppointmentID','ScheduledDay','AppointmentDay','No-show']
features = [f for f in data.columns if f not in omit_columns]   # I purposely left id to demonstrate overfitting
X = data.loc[:,features]
y = np.array(data['No-show'].replace(to_replace=['Yes','No'],value=[1,0]))
```

lets split our data to train and validation (we're leaving a part of the data aside to examine our prediction accuracy)

```python
from sklearn.model_selection import LeavePGroupsOut
lpgo = LeavePGroupsOut(n_groups=10000)
# print lpgo.get_n_splits(groups=data.PatientId,X=X,y=y)
for train_index, test_index in lpgo.split(X, y, data.PatientId):
#     print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.loc[train_index,:], X.loc[test_index,:]
    y_train, y_test = y[train_index], y[test_index]
    break
#     print(X_train, X_test, y_train, y_test)
```

```python
print X_train.shape
print X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix
dtc = DecisionTreeClassifier(max_depth=2,min_samples_split=2,random_state=12345)
dtc.fit(X_train,y_train)
tree_train_pred = dtc.predict(X_train)
print 'train results:'
print(confusion_matrix(y_pred=tree_train_pred,y_true=y_train))
sns.heatmap(confusion_matrix(y_pred=tree_train_pred,y_true=y_train),cmap='Greens',xticklabels=range(2),yticklabels=range(2))
print('classification report results:\r\n' + classification_report(y_pred=tree_train_pred,y_true=y_train))

tree_pred = dtc.predict(X_test)
print 'validation results:'
print(confusion_matrix(y_pred=tree_pred,y_true=y_test))
sns.heatmap(confusion_matrix(y_pred=tree_pred,y_true=y_test),cmap='Greens',xticklabels=range(2),yticklabels=range(2))
print('classification report results:\r\n' + classification_report(y_pred=tree_pred,y_true=y_test))
```

```python
omit_columns = ['AppointmentID','ScheduledDay','AppointmentDay','No-show','PatientId']
features = [f for f in data.columns if f not in omit_columns]   # I purposely left id to demonstrate overfitting
X = data.loc[:,features]
y = np.array(data['No-show'].replace(to_replace=['Yes','No'],value=[1,0]))
```

```python
from sklearn.model_selection import LeavePGroupsOut
lpgo = LeavePGroupsOut(n_groups=10000)
# print lpgo.get_n_splits(groups=data.PatientId,X=X,y=y)
for train_index, test_index in lpgo.split(X, y, data.PatientId):
#     print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.loc[train_index,:], X.loc[test_index,:]
    y_train, y_test = y[train_index], y[test_index]
    break
#     print(X_train, X_test, y_train, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix
dtc = DecisionTreeClassifier(max_depth=300,min_samples_split=2,random_state=12345)
dtc.fit(X_train,y_train)
tree_train_pred = dtc.predict(X_train)
print 'train results:'
print(confusion_matrix(y_pred=tree_train_pred,y_true=y_train))
sns.heatmap(confusion_matrix(y_pred=tree_train_pred,y_true=y_train),cmap='Greens',xticklabels=range(2),yticklabels=range(2))
print('classification report results:\r\n' + classification_report(y_pred=tree_train_pred,y_true=y_train))

tree_pred = dtc.predict(X_test)
print 'validation results:'
print(confusion_matrix(y_pred=tree_pred,y_true=y_test))
sns.heatmap(confusion_matrix(y_pred=tree_pred,y_true=y_test),cmap='Greens',xticklabels=range(2),yticklabels=range(2))
print('classification report results:\r\n' + classification_report(y_pred=tree_pred,y_true=y_test))
```

```python
pd.DataFrame({'feature':X_train.columns,'importance':dtc.feature_importances_}).set_index(['feature']).sort_values(
    ['importance']).iloc[:15,:].plot(kind='barh')
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_jobs=4,n_estimators=300,min_samples_split=2)
rfc.fit(X_train,y_train)
rfc_train_pred = rfc.predict(X_train)
print 'train results:'
print(confusion_matrix(y_pred=rfc_train_pred,y_true=y_train))
sns.heatmap(confusion_matrix(y_pred=rfc_train_pred,y_true=y_train),cmap='Greens',xticklabels=range(2),yticklabels=range(2))
print('classification report results:\r\n' + classification_report(y_pred=rfc_train_pred,y_true=y_train))

rfc_pred = rfc.predict(X_test)
print 'validation results:'
print(confusion_matrix(y_pred=rfc_pred,y_true=y_test))
sns.heatmap(confusion_matrix(y_pred=rfc_pred,y_true=y_test),cmap='Greens',xticklabels=range(2),yticklabels=range(2))
print('classification report results:\r\n' + classification_report(y_pred=rfc_pred,y_true=y_test))
```

```python
import xgboost as xgb

dtrain = xgb.DMatrix(data=X_train,label=y_train) #xgb classes starts from zero
dval = xgb.DMatrix(data=X_test,label=y_test) #xgb classes starts from zero
watchlist = [(dval,'eval'), (dtrain,'train')]

xgb_params = {
    'eta': 0.05,
    'max_depth': 7,
    'subsample': 0.9,
    'colsample_bytree': 0.9,
    'colsample_bylevel': 0.7,
    'alpha':0.1,
    'objective': 'binary:logistic',
#     'objective': 'multi:softmax',
    'eval_metric': 'auc',
#     'eval_metric': 'mlogloss',
    'watchlist':watchlist,
    'print_every_n':5,
    'min_child_weight':2
}

bst = xgb.train(params=xgb_params,dtrain=dtrain,num_boost_round=400)
xgb_pred = bst.predict(dval)
thresh = 0.3
print(confusion_matrix(y_pred=(xgb_pred>thresh)*1,y_true=y_test))
sns.heatmap(confusion_matrix(y_pred=(xgb_pred>thresh)*1,y_true=y_test),cmap='Greens',xticklabels=range(2),yticklabels=range(2))
print('classification report results:\r\n' + classification_report(y_pred=(xgb_pred>thresh)*1,y_true=y_test))
```
